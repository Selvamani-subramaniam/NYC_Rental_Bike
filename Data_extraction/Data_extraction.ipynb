```python
from google.colab import drive
import os
import pandas as pd
import zipfile

# Mount Google Drive
drive.mount('/content/drive', force_remount=True)
#!fusermount -u <mount_point>
# Specify the path to the folder containing your zip files in Google Drive
zip_folder_path = '/content/drive/MyDrive/IE434_Deep_Learning/Data_extraction/Raw_data/Zips/'

# Specify the path to the folder where you want to extract the CSV files
extracted_folder_path = '/content/drive/MyDrive/IE434_Deep_Learning/Data_extraction/Raw_data/extracted_csvs/'

# Create the extracted folder if it doesn't exist
os.makedirs(extracted_folder_path, exist_ok=True)

# Get a list of all zip files in the folder
zip_files = [f for f in os.listdir(zip_folder_path) if f.endswith('.zip')]

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each zip file, extract its contents, and read the CSV files
for zip_file in zip_files:
    zip_file_path = os.path.join(zip_folder_path, zip_file)
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extracted_folder_path)

    # Get a list of all CSV files in the extracted folder
    csv_files = [f for f in os.listdir(extracted_folder_path) if f.endswith('.csv')]

csv_file_path = [os.path.join(extracted_folder_path, csv_file) for csv_file in os.listdir(extracted_folder_path) if csv_file.endswith('.csv')]

# Initialize an empty list to store DataFrames
dfs = []

# Loop through CSV files and read them into DataFrames
for csv_file in csv_file_path:
    df = pd.read_csv(csv_file)
    dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)
```

```python
def encode_and_bind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]])
    res = pd.concat([original_dataframe, dummies], axis=1)
    res = res.drop([feature_to_encode], axis=1)
    return(res)
```

```python
features_to_encode = columns_select = ['rideable_type',
       'start_station_id','member_casual']

for feature in features_to_encode:
    df3 = encode_and_bind(df3, feature)
```
